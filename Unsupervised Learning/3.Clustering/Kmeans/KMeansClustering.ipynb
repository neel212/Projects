```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_set = pd.read_csv('Amazon_data_set_cluster.csv')

X = data_set.iloc[:, 3:5].values

print(X)

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    k_means = KMeans(n_clusters=i, init='k-means++', random_state=21)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('WCSS via elbow method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS Values')
plt.show()

k_means = KMeans(n_clusters=4, init='k-means++', random_state=21)
k_means.fit(X)

y_means = k_means.predict(X)

print(y_means)

plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s=100, color='red', label='Cluster 1')
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s=100, color='yellow', label='Cluster 2')
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s=100, color='blue', label='Cluster 3')
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], s=100, color='green', label='Cluster 4')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=50, color='black', label='Centroids')
plt.title('Clusters of Amazon Users')
plt.xlabel('Annual income in INR')
plt.ylabel('Purchase Rating')
plt.legend(loc='upper right' , fontsize = 5)  # To show labels
plt.show()
#In the context of the K-Means clustering algorithm, the init parameter specifies 
#the method used for initializing the initial cluster centroids. 
#The choice of initial centroids can significantly impact the convergence and quality of the K-Means clustering results.
#'k-means++': This is the default and most commonly used method for initializing cluster centroids. 
#It spreads out the initial centroids to be well-distributed across the data points. 
#It helps in achieving faster convergence and more reliable clustering results.
```
