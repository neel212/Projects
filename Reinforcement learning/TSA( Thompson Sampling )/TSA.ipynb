```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
data_set = pd.read_csv('Upper Confidence Bound Facebook ads.csv')
```

```python
T = 1500
num_ads = 10
ads_selected = []
no_of_reward_1  = [0]*num_ads
no_of_reward_0 = [0]*num_ads
total_rewards = 0

for n in range(0,T):
    ad = 0
    max_random = 0
    for i in range(0,10):
        random_beta = random.betavariate(no_of_reward_1[i]+1, no_of_reward_0[i]+1)
        if(random_beta>max_random):
            max_random = random_beta
            ad = i
        ads_selected.append(ad)
        reward = data_set.values[n,ad]
        if(reward == 1):
            no_of_reward_1[ad] += 1
        else:
            no_of_reward_0[ad] += 1
        total_rewards += reward
```

```python
plt.hist(ads_selected)
plt.title('Number of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad gets selected.')
plt.show()
```
