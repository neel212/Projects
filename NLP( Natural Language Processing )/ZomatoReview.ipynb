```python
import numpy as np
import pandas as pd
```

```python
data_set = pd.read_csv('Zomato_Review.txt', delimiter = '\t', quoting = 3 )
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
#Certainly! The code you've provided appears to be a Python script that uses the Natural Language Toolkit (NLTK) library to perform text preprocessing tasks. Let me explain each line of the code step by step:

# 1. `import re`: This line imports the Python `re` module, which stands for "regular expressions." Regular expressions are used for pattern matching and manipulation of strings. In this code, `re` will be used for text processing tasks.

# 2. `import nltk`: This line imports the Natural Language Toolkit (NLTK), a popular Python library for natural language processing and text analysis.

# 3. `nltk.download('stopwords')`: This line downloads the NLTK dataset for stopwords. Stopwords are common words (e.g., "the," "and," "is") that are often removed from text during text preprocessing tasks because they typically do not carry much meaning.

# 4. `from nltk.corpus import stopwords`: This line imports the `stopwords` module from the NLTK corpus. The `stopwords` module provides a list of common stopwords for various languages.

# 5. `from nltk.stem.porter import PorterStemmer`: This line imports the `PorterStemmer` class from the NLTK library. The Porter Stemmer is a widely used algorithm for stemming, which reduces words to their root or base form. For example, it can convert words like "running," "ran," and "runner" to the common root "run."

# Each line of code in this script imports necessary libraries and resources for text preprocessing tasks, setting the stage for further text analysis or processing.
```

```python
corpus = []
for i in range(0,1000):
    zomato_ratings = re.sub('[^a-zA-Z]', ' ' , data_set['Review'][i])
    print(zomato_ratings)
    print()
    zomato_ratings = zomato_ratings.lower()
    print(zomato_ratings)
    print()
    zomato_ratings = zomato_ratings.split()
    print(zomato_ratings)
    print()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    zomato_ratings = [ps.stem(word) for word in zomato_ratings if not word in set(all_stopwords)]
    print(zomato_ratings)
    print()
    zomato_ratings = ' '.join(zomato_ratings)
    print(zomato_ratings)
    corpus.append(zomato_ratings)
```

```python
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2000)
X = cv.fit_transform(corpus).toarray()
y = data_set.iloc[:,-1].values
print(X)
print()
print(y)
# X = cv.fit_transform(corpus).toarray(): This line fits the CountVectorizer on the corpus data and transforms 
# the text data into a numerical feature matrix X. toarray() converts the sparse matrix into a dense NumPy array. 
# Each row in X represents a document (from the corpus), and each column represents a unique word or token. 
# The values in X represent the counts of each word in the respective documents.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
Classifier = GaussianNB()
Classifier.fit(X_train,y_train)
```

```python
y_predict = Classifier.predict(X_test)
```

```python
print(np.concatenate((y_predict.reshape(len(y_predict),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, r2_score
```

```python
cm = confusion_matrix(y_test, y_predict)
print(cm)
print()
print(accuracy_score(y_test, y_predict))
print()
print(r2_score(y_test, y_predict))
```
