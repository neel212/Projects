```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("RBC in humans polynomial regression.csv")
```

```python
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
# Training the polynomial regression model for the whole data set
from sklearn.preprocessing import PolynomialFeatures # trying to preprocess data in polynomial format.
polynomial_reg = PolynomialFeatures(degree = 6) #AX1^1+AX2^2+...+AnXn^n (n==4).
x_poly = polynomial_reg.fit_transform(X) # Transform preprocessed data into polynomial format degree=6).
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x_poly,y)
```

```python
# Visualizing why linear regression is bad for polynomial data.
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X),color='b')
plt.xlabel('Age')
plt.ylabel('RBC', fontsize = 18)
plt.show()
```

```python
# Visualising polynomial regression
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, linear_reg.predict(x_poly),color='b')
plt.xlabel('Age')
plt.ylabel('RBC', fontsize = 18)
plt.show()
```
